```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html = requests.get("https://en.wikipedia.org/wiki/Red_Bull_BC_One").text
```

```python
soup = BeautifulSoup(html)
```

```python
tables = soup.select("table.wikitable")
len(tables)
```

```python
tables[3]
```

```python
table_2023 = tables[3]
table_2023.text[:500]
```

```python
table_header_row = table_2023.select("th")
len(table_header_row)
```

```python
table_headers = [ table_header_names.text.strip() for table_header_names in table_2023.select("th")[:2] ]
table_headers
```

```python
table_2023_rows = []
for i in range(1, len(table_2023.select("tr"))):
    table_2023.select("tr")[i]
    table_2023_rows.append(table_2023.select("tr")[i])
len(table_2023_rows)
```

```python
table_2023_cells = [
    [ table_2023_cell.text.strip() for table_2023_cell in table_2023_row.select("td")[:2] ]
    for table_2023_row in table_2023_rows
]

table_2023_cells
```

```python
og_df = pd.DataFrame(table_2023_cells, columns=table_headers)
og_df
```

```python
for table_2023_row in table_2023_rows:
    country_2023 = table_2023_row.a.get('title')
    print(country_2023)
```

```python
countries_2023 = []
for table_2023_row in table_2023_rows:
    countries_2023.append(table_2023_row.a.get('title'))

og_df['Country'] = countries_2023
og_df
```

```python
og_df['Year'] = '2023'
og_df
```
